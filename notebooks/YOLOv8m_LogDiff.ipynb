# Entrenar modelo

```python
import shutil
import os

# Nombre del experimento
nombre_experimento = 'yolo_barras_v8m_LogDiff'

# Ruta donde se guarda por defecto
ruta_modelo = f'runs/detect/{nombre_experimento}'

# Verificar y eliminar si ya existe
if os.path.exists(ruta_modelo):
    print(f"Eliminando modelo anterior en: {ruta_modelo}")
    shutil.rmtree(ruta_modelo)

# Entrenar modelo
from ultralytics import YOLO

model = YOLO('yolov8m.pt')

model.train(
    data='/content/drive/MyDrive/MNA_ProyectoIntegrador/galaxy_dataset_2/data.yaml',
    imgsz=640,
    epochs=100,
    batch=32,
    device=0,  # A100
    name=nombre_experimento,
    conf=0.25,
    iou=0.45,
    patience=8,
    cache=True
)
```

```python
# Copiar el mejor modelo a Drive
import shutil

ruta_modelo = "/content/runs/detect/yolo_barras_v8m_LogDiff/weights/best.pt"
destino_drive = "/content/drive/MyDrive/MNA_ProyectoIntegrador/modelos_yolo/best_v8m_LogDiff.pt"

os.makedirs(os.path.dirname(destino_drive), exist_ok=True)
shutil.copy(ruta_modelo, destino_drive)

print(" Modelo copiado a tu Google Drive.")
```
